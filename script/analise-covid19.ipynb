### Introdução, Contexto e Objetivo

A **COVID-19** é uma doença infecciosa causada pelo Coronavírus. É transmitida principalmente por meio de gotículas de pessoas infectadas. A gravidade dos sintomas varia muito de pessoa para pessoa. Até o presente momento, observa-se que cerca de 80% dos casos confirmados são assintomáticos e rápidos. A maioria das pessoas que se encaixam nesse grupo, se recuperam sem nenhuma sequela. No entanto, 20% das pessoas apresentam infecções graves e precisam de oxigênio ou auxílio médico. </br></br>
Em 31 de dezembro de 2019, a Organização Mundial da Saúde (OMS) foi alertada sobre vários casos de pneumonia na cidade de Wuhan na China. Tratava-se de um novo tipo de Coronavírus que não havia sido identificado antes em seres humanos. </br></br>
Uma semana depois, em 7 de janeiro de 2020, as autoridades chinesas confirmaram que haviam identificado um novo tipo de Coronavírus. Logo o novo SARS-CoV2 se espalhou rapidamente pelo mundo. Em 30 de janeiro de 2020, a OMS declarou que o surto do novo Coronavírus constituía em uma Emergência de Saúde Pública de importância internacional. Essa decisão buscou aprimorar a coordenação, a cooperação e a solidariedade global para interromper a propagação do vírus. </br></br>
Atualmente existem quatro vacinas contra a doença que já receberam autorização da Agência Nacional de Vigilância Sanitária (Anvisa) para uso no Brasil: CoronaVac, AstraZeneca, Pfizer e Janssen. Além disso, em maio de 2023, a OMS decretou o fim da emergência pandêmica. </br></br>

````python
df_casos
````

````python
visao_geral_casos = df_casos.groupby(['date', 'state']).agg({'confirmed':'sum'}).reset_index()

visao_geral_casos['mes'] = visao_geral_casos['date'].dt.strftime('%Y-%m')

visao_geral_casos
````

````python
casos_data_estado = visao_geral_casos.groupby(['mes', 'state']).agg({'date':'max'}).reset_index()

casos_data_estado
````

````python
ultimas_atualizacoes_mensais_estados = pd.merge(casos_data_estado, visao_geral_casos, how = 'inner', on = ['state', 'date'])

ultimas_atualizacoes_mensais_estados
````

````python
casos_confirmados_mensais = ultimas_atualizacoes_mensais_estados.groupby(['mes_x']).agg({'confirmed':'sum'}).reset_index()

casos_confirmados_mensais
````

````python
fig_casos_mensais = px.line(casos_confirmados_mensais, x='mes_x', y ='confirmed', labels={
    'mes_x': 'Mês Ocorrência',
    'confirmed': 'Total Confirmados'})
    
fig_casos_mensais.update_layout(title={
    'text' : 'Quantidade de Casos Confirmados por Mês',
    'y': 0.95,
    'x': 0.5
})
    
fig_casos_mensais
````

````python
ultima_atualizacao_estado = ultimas_atualizacoes_mensais_estados.groupby(['state']).agg({'date':'max'}).reset_index()

casos_confirmados_estados = pd.merge(ultimas_atualizacoes_mensais_estados, ultima_atualizacao_estado, how = 'inner', on = ['state', 'date'])

casos_confirmados_estados
````

````python
fig_casos_estados = px.bar(casos_confirmados_estados, x='state', y ='confirmed', labels={
    'state': 'Estado',
    'confirmed': 'Total Confirmados'})

fig_casos_estados.update_layout(title={
    'text' : 'Quantidade de Casos Confirmados por Estado',
    'y': 0.95,
    'x': 0.5
})

fig_casos_estados
````

### Conclusão

Em resumo, a pandemia da Covid-19 marcou uma época sem precedentes na história recente, causando impactos profundos em todas as esferas da sociedade global. Ao longo deste trabalho, buscamos responder quais os principais fatores que implicaram para a rápida propagação da doença. Se existe algum período em que os casos tendem a aumentar e, por último, tendo em vista que a vacina não é obrigatória, qual o perfil do público vacinado.</br></br>

Ao longo dessa pesquisa, encontramos dados que mostram que um dos principais motivos da propagação da doença é o contato humano próximo. Isso inclui reuniões sociais, ambientes de trabalho, transporte público e espaços lotados onde o distanciamento físico não é mantido adequadamente, não é atoa que São Paulo (estado com maior população do Brasil), foi o estado com maior número de casos confirmados. Além disso, o fato da forma de transmissão do vírus ser através do ar e existir pessoas infectadas que são assintomáticas (Não demonstram nenhum efeito de estarem contaminadas), são outros dos principais fatores da propagação.</br></br>

Neste trabalho identificamos um grande aumento no período de Março de 2021, até seu pico em Agosto de 2021. Esse mesmo pico de casos crescentes aconteceu praticamente no mesmo período de 2020 (Junho a Setembro - Cerca de dois milhões de casos por mês). Apesar do indicador se comportar de maneira semelhante em ambos os períodos, não pode-se inferir que isto esteja diretamente interligado ao período do ano ou até com questões climáticas. Para indicar ou não alguma relação com uma época do ano específica, seria necessário um conjunto de dados históricos maior.</br></br>

Além disso, onseguimos identificar que o maior número de vacinados foram pessoas do sexo feminino e de raça/cor branca, pessoas da raça/cor preta foram as menos vacinadas de acordo com o número total.

Após a análise completa desses dados, percebemos que os maiores picos da doença foram nos anos de 2020 e 2021, e que também os casos passaram a diminuir após os usos de máscaras e vacinações, acreditamos que a melhor recomendação para possíveis crises futuras, desta ou de uma potencial nova doença, é enquanto não tivermos informações ou dados da doença, seguir as indicações da OMS. Em resumo, a pandemia de Covid-19 trouxe desafios sem precedentes para a sociedade global. Porém, também destacou a importância da colaboração internacional, da ciência e da resiliência humana.
